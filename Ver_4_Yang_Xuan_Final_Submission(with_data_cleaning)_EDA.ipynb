```python
# Read multiple sheet from UNdata excel sheet into a dictionary
UNdata_dict_df = pd.read_excel("UN_MigrantStockTotal_2015.xlsx",
                       sheet_name=['Table 1', 'Table 2', 'Table 3', 'Table 4', 'Table 5', 'Table 6'])
```

```python
# Create 6 dataframe from the dictionary
df1 = UNdata_dict_df.get('Table 1')
df2 = UNdata_dict_df.get('Table 2')
df3 = UNdata_dict_df.get('Table 3')
df4 = UNdata_dict_df.get('Table 4')
df5 = UNdata_dict_df.get('Table 5')
df6 = UNdata_dict_df.get('Table 6')
```

```python
# Read in df1 and check out its characteristics
df1
```

```python
# Find out what columns are in Table 1
df1.columns
```

```python
# Row 1 to Row 16 are texts, which are irrelavent to the data
# We also dropped the current column titles for Table. We are going to assign new headers to each column later.
# Drop Row 1 through Row 15
df1_processed = df1.drop(df1.index[:15])
```

```python
# Check out the result of the dropped table
print(df1_processed)
```

```python
# We successfully dropped the texts, now we want to deal with the unnamed columns.
df1_processed.head(10)
```

```python
# Double check the total number of the columns
df1_processed.shape
```

```python
# Rename the column names that will be easier to split in later steps
# As we did in class, I will use year+sex as a set to name the columns first, then I can split year from sex.
# I will use B to represent both sex, F to represent female, M to represent male
df1_processed.rename(columns={df1.columns[0]:"Order", df1.columns[1]:"Major area, region, country or area of destination",
                    df1.columns[2]:"Notes",df1.columns[3]:"Country Code",df1.columns[4]:"Type of Data",
                    df1.columns[5]:"1990B",df1.columns[6]:"1995B",df1.columns[7]:"2000B",df1.columns[8]:"2005B",df1.columns[9]:"2010B",
                    df1.columns[10]:"2015B",df1.columns[11]:"1990M",df1.columns[12]:"1995M",df1.columns[13]:"2000M",df1.columns[14]:"2005M",df1.columns[15]:"2010M",df1.columns[16]:"2015M",
                    df1.columns[17]:"1990F",df1.columns[18]:"1995F",df1.columns[19]:"2000F",df1.columns[20]:"2005F",df1.columns[21]:"2010F",df1.columns[22]:"2015F",}, inplace = True)
```

```python
# Check the results of renaming the headers, all the columns look correct now. 
# Then we move on to splitting columns (column names) with multiple variables
df1_processed.head(5)
```

```python
# Check all column names
df1_processed.columns
```

```python
# Before I do any formal cleaning on the dataset, I would like to drop "Notes" column first
df1_processed = df1_processed.drop(["Notes"], axis=1)
```

```python
# Check if I correctly dropped the "Notes" column
df1_processed.head()
```

```python
# Principle 2 of Tidy Data: Multiple variabes should not be stored in one column
# Sex and year are stored in one column, so we need to split, and melt the table and seperate the information
# I will keep "Order", "Major area", "country code", and "type of data" constant.
# The value is "International migrant stock at mid-year", as mentioned in the EXCEL sheet
df1_melted = (df1_processed.melt(id_vars=['Order','Major area, region, country or area of destination','Country Code', 'Type of Data'], 
                                 value_name = "International Migrant Stock at Mid-year"))
```

```python
# Check if I melted the table correctly
# I successfully splitted year+sex from "international migrant stock at mid-year"
df1_melted.head(10)
```

```python
# Chech the tail of the table
df1_melted.tail(10)
```

```python
# Check the columns of the table
# A new variable called "variable" is created with the 'melt' function
df1_melted.columns
```

```python
# Then I want to rename the "variable" variabe to year
# I will use a lambda function to split a "SEX" column after renaming the "Year" variable
df1_melted = df1_melted.rename(columns={df1_melted.columns[4]:"Year"})
```

```python
# Chech the result of renaming the Year variable
df1_melted
```

```python
# Now I will use a lambda function to slice "Year" into "Year" and "Sex"
df1_melted = df1_melted.assign(Sex = lambda x: x.Year.str[4].astype(str), Year = lambda x: x.Year.str[:4].astype(str))
```

```python
# Chech if the lambda function worked or not
df1_melted.head(10)
```

```python
# Check the tail of the table as well
df1_melted.tail(10)
```

```python
# Split the Major area column into 'Regions' and 'Major area, region, country or area of destination' 
last_region = None
regions = []
kept_rows = []
for row in df1_melted.iterrows():
  if row[1][2] >= 900:
    last_region = row[1]["Major area, region, country or area of destination"]
    continue
  regions.append(last_region)
  kept_rows.append(row[0])
df1_melted = df1_melted.iloc[kept_rows]
df1_melted.insert(0, "Regions", regions, allow_duplicates=True)
```

```python
# Check if we correctly split the two columns.
df1_melted.head()
```

```python
# Rename "major area" to "Country of destination"
df1_melted = df1_melted.rename(columns={df1_melted.columns[2]:"Country of destination"}) 
```

```python
# Check if correctly rename the column
df1_melted.head(20)
```

```python
# Drop columns 'Order' and 'Type of Data'
df1_melted = df1_melted.drop(["Order","Type of Data"], axis=1)
```

```python
# Check if drop the correct columns
df1_melted.head(10)
```

```python
# Replace the capital letter "B", "F", "M" with "Both","Female","Male"
df1_melted = (df1_melted.replace(to_replace =["B","M","F"],value =["Both","Male","Female"]))
```

```python
# Check if change the content properly
df1_melted.head()
```

```python
# Last step of cleaning the first table
# Save Table 1 as .csv file
df1_melted.to_csv("Table1.csv")
```
